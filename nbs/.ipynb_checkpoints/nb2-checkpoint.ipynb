# Evaluation and comparing ML models

## Imputing missing values

Before proceeding to the evaluation of ML models it is necessary to fill in any missing values. [This paper](http://www.stat.columbia.edu/~gelman/arm/missing.pdf) discusses the process of imputation in depth.

```python
# create imputer. Fill missing values with median
imputer = impute.SimpleImputer(strategy='median')

# tain on training features
imputer.fit(train_features)

# transform both training and testing data
X = imputer.transform(train_features)
X_test = imputer.transform(test_features)
```

Checking for:
- no missing values
- no infinite values

```python
# check for missing values
print(f'missing train values:  {np.isnan(X).sum()}')
print(f'missing test values:   {np.isnan(X_test).sum()}')

# check for infinite values
print(f'infinite train values: {(~np.isfinite(X)).sum()}')
print(f'infinite test values:  {(~np.isfinite(X_test)).sum()}')
```

## Scaling features

Features are currently expressed in different units. Altgorithms that take into consideration the Eucledian distance between observations will be affected.

We will scale all units to ensure all algorithms can be compared.

Two main types of scaling:
- Scaling to a mean of 0 and standard deviation of 1, also known  as standardisation.
- Scaling to a range by substracting the minimum value and dividing by the range, known as normalisation.

```python
# create scaler object. Normalise with a range of 0 to 1
scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
```

```python
# fit in train data only to avoid information leakage
scaler.fit(X)
```

```python
# transform train and test data
X = scaler.transform(X)
X_test = scaler.transform(X_test)
```

```python
# transform target variable to a 1D array
y=np.array(train_labels).reshape((-1,))
y_test=np.array(test_labels).reshape((-1,))
```

```python
# save features and target to memory
with open('../data/features_train.p','wb') as f:
    pickle.dump(X,f)
    
with open('../data/features_test.p','wb') as f:
    pickle.dump(X_test,f)
    
with open('../data/target_train.p','wb') as f:
    pickle.dump(y,f)

with open('../data/target_test.p','wb') as f:
    pickle.dump(y_test,f)
```

## Models evaluation

The following models will be evaluated:
- Linear Regression
- SVM Regressor
- Random Forests Regressor
- Gradient Boosting Regressor
- K-Nearest Neighbours Regressor

```python
# evaluate baseline metric = mean absolute errror
def mae(y_true, y_pred):
    return np.mean(abs(y_true-y_pred))

# train and evaluate model on test set
def fit_and_evaluate(model, X, X_test, y, y_test):
    model.fit(X,y)
    model_pred = model.predict(X_test)
    model_mae = mae(y_test, model_pred)
    return model_mae  
```

```python
lr = linear_model.LinearRegression()
lr_mae = fit_and_evaluate(lr, X, X_test, y, y_test)
```

```python
svmr = svm.SVR(C=1000, gamma=0.1)
svmr_mae = fit_and_evaluate(svmr, X, X_test, y, y_test)
```

```python
random_forest = ensemble.RandomForestRegressor(random_state=60)
random_forest_mae = fit_and_evaluate(random_forest, X, X_test, y, y_test)
```

```python
gradient_boost = ensemble.GradientBoostingRegressor(random_state=60)
gradient_boost_mae = fit_and_evaluate(gradient_boost, X, X_test, y, y_test)
```

```python
knn = neighbors.KNeighborsRegressor(n_neighbors=10)
knn_mae = fit_and_evaluate(knn, X, X_test, y, y_test)
```

```python
mae_results = [
    {'algo':'linear_regression','mae':lr_mae},
    {'algo':'svm','mae':svmr_mae},
    {'algo':'random_forest','mae':random_forest_mae},
    {'algo':'gradient_boost','mae':gradient_boost_mae},
    {'algo':'knn', 'mae':knn_mae}
]

mae_results = pd.DataFrame(mae_results)

mae_results.sort_values(by='mae', ascending=False).plot(x='algo', y='mae', 
                                                        kind='barh', color='r', 
                                                        edgecolor='b');
plt.vlines(x=24.39, ymin=0, ymax=np.max(mae_results.mae),
           color='r')
plt.text(x=22.39,y=0,s="baseline")
plt.title('MAE comparison')
plt.xlabel('MAE')
plt.ylabel('Algorithm');
```

We can conclude from this results that ML is suitable to solve this problem as all models performed better than the baseline.

# Model Optimisation

Given the results obtained in our model evaluation, we will optimise the gradient boost algorithm using sklearn's RandomSearchCV and k-fold cross validation.

```python
# loss functions to optimise
loss = ['ls', 'lad', 'huber']

# number of trees
n_estimators = [100,500,900,1100,1500]

# max depth of each tree
max_depth = [2,3,5,10,15]

# min samples per leaf
min_samples_leaf = [1, 2, 4, 6, 8]

# min samples to split a node
min_samples_split = [2, 4, 6, 10]

# max features to consider for making splits
max_features = ['auto', 'sqrt', 'log2', None]

# hyperparameters grid
hyperparameter_grid = {'loss': loss,
                       'n_estimators': n_estimators,
                       'max_depth': max_depth,
                       'min_samples_leaf': min_samples_leaf,
                       'min_samples_split': min_samples_split,
                       'max_features': max_features}
```

```python
# creating the model
model = ensemble.GradientBoostingRegressor(random_state=42)
```

```python
# setting random search with 4-fold cross validation
random_cv = model_selection.RandomizedSearchCV(estimator=model,
                                               param_distributions=hyperparameter_grid,
                                               n_iter=25,scoring='neg_mean_absolute_error',
                                               cv=4, n_jobs=-1, verbose=1, 
                                               return_train_score = True,random_state=42)
```

```python
random_cv.fit(X, y)
```

```python
with open('../data/random_cv_clf.p','wb') as f:
    pickle.dump(random_cv,f)
```

```python
random_results = pd.DataFrame(random_cv.cv_results_).sort_values(by='mean_test_score', ascending=False)
random_results.head(10)
```

```python
random_cv.best_estimator_
```

The best estimator derived from the random cross validator provides a good idea of the hyperparameters that will best suit the problem.

To continue the optimisation of the model, we will focus on the number of trees in the forest (`n_estimators`)

```python
trees = {'n_estimators': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800]}

model = ensemble.GradientBoostingRegressor(loss = 'lad', max_depth = 5,
                                           min_samples_leaf = 6,
                                           min_samples_split = 6,
                                           max_features = None,
                                           random_state = 42)

grid_search = model_selection.GridSearchCV(estimator = model, param_grid = trees,
                                           scoring='neg_mean_absolute_error', verbose=1,
                                           n_jobs=-1, return_train_score=True)
```

```python
grid_search.fit(X,y)
```

```python
with open('../data/n_estimators_grid_search_clf.p','wb') as f:
    pickle.dump(grid_search, f)
```

```python
results = pd.DataFrame(grid_search.cv_results_)
```

```python
results.info()
```

```python
plt.figure(figsize=(8,8))
plt.style.use('fivethirtyeight')
plt.plot(results.param_n_estimators,-1* results.mean_test_score, label='Test Error')
plt.plot(results.param_n_estimators,-1* results.mean_train_score, label='Train Error')
plt.xlabel('Number of Estimators')
plt.ylabel('Mean Absolute Error')
plt.title('Performance vs. Number of Trees')
plt.legend();
```

```python
results.sort_values(by='mean_test_score', ascending=False)[['mean_test_score','mean_train_score','params']]
```

```python
grid_search.best_estimator_
```

The model appears to be overfitting to the training set. For the model to generalise better to the test data I will test increasing the number of samples per leaf, reducing the max depth for each tree, and reducing the number of trees.

The testing set reduced the error up until about 300 estimators at which point it appears to plateau.

I'll explore the effect of the depth of the trees on the model performance.

```python
# max depth of each tree
max_depth = np.linspace(1,30,30,endpoint=True)

model = ensemble.GradientBoostingRegressor(loss = 'lad', 
                                           n_estimators = 750,
                                           min_samples_leaf = 6,
                                           min_samples_split = 6,
                                           max_features = None,
                                           random_state = 42)

grid_search = model_selection.GridSearchCV(estimator = model, param_grid = {'max_depth':max_depth},
                                           scoring='neg_mean_absolute_error', verbose=1,
                                           n_jobs=-1, return_train_score=True)
```

```python
grid_search.fit(X,y)
```

```python
with open('../data/max_depth_grid_search_clf.p','wb') as f:
    pickle.dump(grid_search, f)
```

```python
results = pd.DataFrame(grid_search.cv_results_)
```

```python
plt.figure(figsize=(8,8))
plt.style.use('fivethirtyeight')
plt.plot(results.param_max_depth,-1* results.mean_test_score, label='Test Error')
plt.plot(results.param_max_depth,-1* results.mean_train_score, label='Train Error')
plt.xlabel('Max Depth')
plt.ylabel('Mean Absolute Error')
plt.title('Performance vs. Max Depth')
plt.legend();
```

```python
grid_search.best_estimator_
```

We can see that large values of max depth are causing our model to overfit. After `max_depth` > 5 the testing set improvement plateaus. I will be trying a low number on both `max_depth` and `n_estimators`

Next I will explore the impact of the number of samples per leaf (`min_samples_leaf`)

```python
# min samples per leaf
min_samples_leaf = np.linspace(1,30,30,endpoint=True).astype(int)

model = ensemble.GradientBoostingRegressor(loss = 'lad', 
                                           n_estimators = 750,
                                           max_depth = 5,
                                           min_samples_split = 6,
                                           max_features = None,
                                           random_state = 42)

grid_search = model_selection.GridSearchCV(estimator = model, param_grid = {'min_samples_leaf':min_samples_leaf},
                                           scoring='neg_mean_absolute_error', verbose=1,
                                           n_jobs=-1, return_train_score=True)
```

```python
grid_search.fit(X,y)
```

```python
with open('../data/min_samples_leaf_grid_search_clf.p','wb') as f:
    pickle.dump(grid_search, f)
```

```python
results = pd.DataFrame(grid_search.cv_results_)
```

```python
plt.figure(figsize=(8,8))
plt.style.use('fivethirtyeight')
plt.plot(results.param_min_samples_leaf,-1* results.mean_test_score, label='Test Error')
plt.plot(results.param_min_samples_leaf,-1* results.mean_train_score, label='Train Error')
plt.xlabel('Min Samples Leaf')
plt.ylabel('Mean Absolute Error')
plt.title('Performance vs. Min Samples Leaf')
plt.legend();
```

```python
grid_search.best_estimator_
```

There wasn't a clear advantage to increasing or decreasing the min samples per leaf, I will select 7 as the curve appears to have a reasonable slope at that level.

# Evaluating final model on test set

```python
# default model 
default_model = ensemble.GradientBoostingRegressor(random_state=42)

# best model
final_model = grid_search.best_estimator_

final_model
```

```python
%%timeit -n 1 -r 5
default_model.fit(X, y)
```

```python
%%timeit -n 1 -r 5
final_model.fit(X, y)
```

```python
default_pred = default_model.predict(X_test)
final_pred = final_model.predict(X_test)

print('Default model performance on the test set: MAE = %0.4f.' % mae(y_test, default_pred))
print('Final model performance on the test set:   MAE = %0.4f.' % mae(y_test, final_pred))
```

```python
# density plot of predictions and values
plt.figure(figsize=(8,8))

sns.kdeplot(final_pred, label = 'final predictions')
sns.kdeplot(default_pred, label = 'default predictions')
sns.kdeplot(y_test, label = 'values')
plt.vlines(x=np.median(y_test), ymin=0, ymax=0.016, alpha=0.3);
```

Both the final and default models predict values closer to the median than the actual values. This could indicate the model tends to predict values closer to the median and does not handle extreme values well.

```python
plt.figure(figsize=(8,8))
residuals=final_pred-y_test
plt.hist(residuals, color = 'red', bins = 20,
         edgecolor = 'black')
plt.xlabel('Error'); plt.ylabel('Count')
plt.title('Distribution of Residuals');
```

```python
plt.figure(figsize=(4,8))
plt.boxplot(residuals);
```

The distributions of residuals is normally distributed with a few noticeable outliers at the lower end.
