```python
# replacing 'Not Available' to np.nan
data = data.replace({'Not Available':np.nan})
```

```python
# convert numeric columns to float
for col in list(data.columns):
    # Select columns that should be numeric
    if ('ft²' in col or 'kBtu' in col or 'Metric Tons CO2e' in col or 'kWh' in 
        col or 'therms' in col or 'gal' in col or 'Score' in col):
        # Convert the data type to float
        data[col] = data[col].astype(float)
```

```python
# number of null values per column
missing_values = data.isnull().sum()

# percentage out of total
pct_missing = missing_values/len(data)

missing_table = pd.concat([missing_values,pct_missing], axis=1)
```

```python
missing_table.rename(columns={0:'Missing Values', 1:'Missing Pct of Total'}, inplace=True)
missing_table[missing_table['Missing Values']!=0].sort_values(by='Missing Pct of Total', ascending=False)
```

```python
# droping colums > 50% missing data
cols_to_drop = list(missing_table[missing_table['Missing Pct of Total']>.5].index)
data.drop(columns = cols_to_drop, inplace=True)
```

## Removing Outliers

Removing outliers. Outliers defined as 
- First Quartile - 3*IQR
- Third Quartile + 3*IQR


```python
# removing outliers from site_eui

fq = data.site_eui.quantile(.25)
tq = data.site_eui.quantile(.75)
iqr = tq-fq

bottom_outliers = fq-(3*iqr)
top_outliers = tq+(3*iqr)

data = data[(data.site_eui>bottom_outliers) &
            (data.site_eui<top_outliers)]
```

## Correlations

```python
data.corr()['score'].sort_values()
```

Strongest negative correlations found with metrics of efficiency where the lower the number the better. To account for non-linear relations, we will calculate correlations with the log and square root transforms.

```python
# create cols with sqrt and log of numeric columns
for c in data.select_dtypes('number').columns:
    # skip target column
    if c=='score':
        next
    else:
        data[f'sqrt_{c}']=np.sqrt(data[c])
        data[f'log_{c}']=np.log(data[c])
```

```python
# one hot encode categorical variables
dummies = pd.get_dummies(data[['Borough', 'Largest Property Use Type']])
```

```python
data=pd.concat([data,dummies], axis=1)
```

```python
features=data.select_dtypes('number').copy()
```

```python
features.dropna(subset=['score'], inplace=True)
```

```python
features.corr()['score'].dropna().sort_values()
```

```python
# add building types to feature set
features['Largest Property Use Type']=data.dropna(subset=['score'])['Largest Property Use Type']
```

```python
# Limit to building types with more than 100 observations
features = features[features['Largest Property Use Type'].isin(types)]
```

```python
# create scatterplot of score vs log source eui
sns.lmplot('site_eui', 'score', data = features,
           hue='Largest Property Use Type', scatter_kws={'alpha':0.8, 's':60},
           fit_reg=False, aspect=1.2, height=12)

# Plot labeling
plt.xlabel("Site EUI")
plt.ylabel('Energy Star Score')
plt.title('Energy Star Score vs Site EUI');
```

### Pairs Plot

```python
# columns to plot
plot_data = features[['score', 'site_eui', 
                      'Weather Normalized Source EUI (kBtu/ft²)', 
                      'log_Total GHG Emissions (Metric Tons CO2e)']]
```

```python
# replace the inf with nan
plot_data = plot_data.replace({np.inf: np.nan, -np.inf: np.nan})
```

```python
# rename columns 
plot_data = plot_data.rename(columns = {'site_eui': 'Site EUI', 
                                        'Weather Normalized Source EUI (kBtu/ft²)': 'Weather Norm EUI',
                                        'log_Total GHG Emissions (Metric Tons CO2e)': 'log GHG Emissions'})
```

```python
# drop na
plot_data.dropna(inplace=True)
```

```python
# function to calculate correlation coefficient between two columns
def corr_func(x, y, **kwargs):
    r = np.corrcoef(x, y)[0][1]
    ax = plt.gca()
    ax.annotate("r = {:.2f}".format(r),
                xy=(.2, .8), xycoords=ax.transAxes,
                size = 20)
```

```python
# create pairgrid object
grid = sns.PairGrid(data = plot_data, height=3)

# upper = scatter plots
grid.map_upper(plt.scatter, color='red', alpha=0.6)

# diagonal = histograms
grid.map_diag(plt.hist, color='red', edgecolor='black')

# bottom = density plot + correlation
grid.map_lower(corr_func);
grid.map_lower(sns.kdeplot, cmap = plt.cm.Reds)

# title for entire plot
plt.suptitle('Pairs Plot of Energy Data', size = 36, y = 1.02);
```
