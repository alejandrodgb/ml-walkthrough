```python
import pandas as pd
import numpy as np
from pathlib import Path
import os
```

```python
path = Path.home()/'Documents/ML-walkthrough'
data_path = Path.home()/'Documents/ML-walkthrough/data'
nbs_path = Path.home()/'Documents/ML-walkthrough/nbs'
```

```python
# read data into dataframe

data = pd.read_csv(data_path/'Energy_and_Water_Data_Disclosure_for_Local_Law_84_2017__Data_for_Calendar_Year_2016_.csv')
```

```python
data.sample(5)
```

```python
data.info()
```

```python
# replacing 'Not Available' to np.nan
data = data.replace({'Not Available':np.nan})
```

```python
# convert numeric columns to float
for col in list(data.columns):
    # Select columns that should be numeric
    if ('ft²' in col or 'kBtu' in col or 'Metric Tons CO2e' in col or 'kWh' in 
        col or 'therms' in col or 'gal' in col or 'Score' in col):
        # Convert the data type to float
        data[col] = data[col].astype(float)
```

```python
# number of null values per column
missing_values = data.isnull().sum()

# percentage out of total
pct_missing = missing_values/len(data)

missing_table = pd.concat([missing_values,pct_missing], axis=1)
```

```python
missing_table.rename(columns={0:'Missing Values', 1:'Missing Pct of Total'}, inplace=True)
missing_table[missing_table['Missing Values']!=0].sort_values(by='Missing Pct of Total', ascending=False)
```

```python
# droping colums > 50% missing data
cols_to_drop = list(missing_table[missing_table['Missing Pct of Total']>.5].index)
data.drop(columns = cols_to_drop, inplace=True)
```
